```python
import requests
import pandas as pd
import re
```

```python
from bs4 import BeautifulSoup
```

```python
URL = 'https://grants.nih.gov/grants/glossary.htm'
```

```python
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
results = soup.find_all('strong', class_='headline2')
```

```python
# https://medium.com/opex-analytics/simple-web-scraping-in-python-90d6fddfaeca
df = pd.DataFrame(columns =['Term', 'Description'])

term = []
description = []

for i in soup.find_all('strong', class_='headline2'):
    term.append(i.get_text())
    next_div = i.find_next(['div', 'p'])
    next_div_text = next_div.get_text().strip().encode("ascii", "replace").decode()
    next_div_text = next_div_text.replace('?', ' ')
    next_div_text = re.sub(' +',' ',next_div_text)
    print(next_div_text.strip())
    #print(next_div.get_text())
    text = []
    #text.append(next_div.get_text())
    text.append(next_div_text)
    print(text)
    description.append(' '.join(text))
    
df['Term'] = term
df['Description'] = description
```
